```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
url='/content/drive/MyDrive/Colab Notebooks/data/ForbesRichestAtheletes(Forbes RichestAthletes1990-2020).csv'
df=pd.read_csv(url)
df.head()
```

##Limpiar

```python
df.drop('S.NO',axis=1,inplace=True)
df
```

```python
cabecera=['Nombre','Nacionalidad','RangoActual','RangoAñoAnterior','Deporte','Anio','ganancias_millones']
df.columns=cabecera
```

```python
df.info()
```

```python
df['Anio']=pd.to_datetime(df['Anio'], format='%Y')
df
```

```python
df.info()
```

```python
#not ranked=9999  none=6666
df['RangoAñoAnterior']=df['RangoAñoAnterior'].fillna('0')
df['RangoAñoAnterior']=df['RangoAñoAnterior'].str.replace('?','0') 
df['RangoAñoAnterior']=df['RangoAñoAnterior'].str.replace('not ranked','9999') 
df['RangoAñoAnterior']=df['RangoAñoAnterior'].str.replace('none','6666') 
#df = df.replace({'RangoAñoAnterior' : { '?' : '0', 'not ranked' : '9999', 'none' : '6666' }})
df.loc[df['RangoAñoAnterior'].str.contains('>'), 
                  'RangoAñoAnterior'] = (df.loc[df['RangoAñoAnterior']
                                                             .str.contains('>')]['RangoAñoAnterior']
                                           .str.strip('>').astype(int)+1).astype(str)
df['RangoAñoAnterior']=df['RangoAñoAnterior'].astype(int)
df.info()
```

## Deportistas cuyo ranking ha subido al menos dos lugares entre 2010 y 2020

```python
df_2010_2020=df.loc[((df['Anio'].dt.year>=2010) & (df['Anio'].dt.year<=2020))]
depMayorRank2=df_2010_2020.loc[(df_2010_2020['RangoActual']-df_2010_2020['RangoAñoAnterior'])>=2]
depMayorRank2.head()
```

## Ganancia mínima y máxima (dentro del dataset) por deporte y por país
